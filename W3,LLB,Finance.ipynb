```python

import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
data=pd.read_csv('telecom_churn.csv')
Clr = data['Churn'].map({False:'blue',True:'orange'})
Clr.head()

fig=plt.figure(figsize=(11,7))
plt.scatter(data['Total intl charge'],data['Account length'],color=Clr,
            s=(data['Account length']+0.05));
plt.xlabel('Total intl charge');
plt.ylabel('Account length')
plt.show()
```
